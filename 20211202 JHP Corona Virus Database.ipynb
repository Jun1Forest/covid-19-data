```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('owid-covid-data.csv')
df
```

```python
def country(country_code):
    df_c = df[df['iso_code']==country_code]
    df_newcases = df_c[['new_cases', 'new_deaths', 'total_deaths']]
    df_newcases.plot()
    plt.title( country_code )
    
    return 
```

```python
country('JPN')
```

```python
country('DEU')
```

```python
country('KOR')
```
